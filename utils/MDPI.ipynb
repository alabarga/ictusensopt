```python
issues = []
```

```python
import time
```

```python
for n in range(1,1448):
    url = f'https://www.mdpi.com/special-issues?page_no={n}'
    print(url)
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    rows = soup.find('table').find_all('tr')
    for row in rows[1:]:
        items = row.find_all('td')
        issue = (items[0].text.strip(),
         items[1].text.strip(),
         items[2].text.strip(),
         items[2].a["href"],
         items[3].text.strip(),
         items[4].text.strip())
        issues.append(issue)
    time.sleep(1)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(issues, columns=['journal', 'section', 'title', 'url', 'fecha', 'articles'])
```
